```python
import pandas as pd

database = pd.read_csv('data_anal.csv')
```

```python
database.head()
```

```python
bulletsVsEnemiesKilled = database.groupby('bulletsUsed')['enemiesKilled'].sum().reset_index()
bulletsVsEnemiesKilled.plot(x='bulletsUsed', y='enemiesKilled', ylabel='Enemies Killed', xlabel='Bullets Used', title='Enemies Killed vs Bullets Used')
```
